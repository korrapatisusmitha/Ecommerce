```sql
CREATE OR REPLACE VIEW ecommerce.gold.fact_transactions_denorm AS (
  SELECT o.*,
  d.year,
  d.month_name,
  d.day_name,
  d.is_weekend,
  d.quarter,
  d.week,
  p.sku,
  p.category_code,
  p.category_name,
  p.brand_code,
  p.brand_name,
  p.color,
  p.size,
  p.rating_count,
  extract(HOUR FROM o.transaction_ts) as hour_of_day,
  c.country,
  c.state,
  c.region
  FROM ecommerce.gold.gld_fact_order_items o 
  JOIN ecommerce.gold.gld_dim_date d ON o.date_id = d.date_id
  JOIN ecommerce.gold.gld_dim_products p ON o.product_id = p.product_id
  JOIN ecommerce.gold.gld_dim_customers c ON o.customer_id = c.customer_id
);
```
